```python
genes = ds.ra['Gene']
```

```python
import pickle
with open("/efs-private/Genecorpus-30M/token_dictionary.pkl", "rb") as fp:
    token_dictionary = pickle.load(fp)
```

```python
from pyensembl import EnsemblRelease
import os
os.environ['PYENSEMBL_CACHE_DIR']='/shared/.cache/pysensembl'
data = EnsemblRelease(110)

def get_gene_id(x):
    try:
        gid = data.gene_ids_of_gene_name(x)
        if isinstance(gid, list):
            return gid[0]
        else:
            return gid
    except ValueError as e:
        print(e)
        print(f"{x} not found in Ensembl")
        return None
```

```python
ensembls = [get_gene_id(x) for x in genes]
```

```python
import numpy as np
gf_match = np.array([True if x in token_dictionary.keys() else False for x in ensembls ])
```
